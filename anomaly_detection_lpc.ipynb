```python
def get_time(r):
    return r['systemtime']

def meanSD(x):
    temp = 0
    for x1 in x:
        temp = temp + x1
    mean = temp/len(x) #mean value
    m=[]
    for y in x:
        k = (y-mean)**2
        m.append(k)
    temp2 = 0
    for x2 in m:
        temp2 = temp2 + x2
    SD = sqrt(temp2/len(x)) #standard deviation
    return mean, SD

def getSpeedForService(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['in_speed']+x['out_speed']

def error(f, x, y):
    return sp.sum((f(x)-y)**2)

def collect_services(reportList):
    result = set()
    for x in reportList:
        for y in x['services']:
            result.add(y['name'])
    return result

def GETfeature_speeds(servicename):
    feature_speeds = []
    for x in reports:
        feature_speeds.append(getSpeedForService(x, servicename))
    return feature_speeds

def lpc_ref(signals, order):
    """Compute the Linear Prediction Coefficients.

    Return the order + 1 LPC coefficients for the signal. c = lpc(x, k) will
    find the k+1 coefficients of a k order linear filter:

      xp[n] = -c[1] * x[n-2] - ... - c[k-1] * x[n-k-1]

    Such as the sum of the squared-error e[i] = xp[i] - x[i] is minimized.

    Parameters
    ----------
    signal: array_like
        input signal
    order : int
        LPC order (the output will have order + 1 items)

    Notes
    ----
    This is just for reference, as it is using the direct inversion of the
    toeplitz matrix, which is really slow"""
    if isinstance(signals, list):
        signals = np.asarray(signals)
    if signals.ndim > 1:
        raise ValueError("Array of rank > 1 not supported yet")
    if order > len(signals):
        raise ValueError("Input signal must have a lenght >= lpc order")

    if order > 0:
        p = order + 1
        r = np.zeros(p, signals.dtype)
        # Number of non zero values in autocorrelation one needs for p LPC
        # coefficients
        nx = np.min([p, signals.size])
        x = np.correlate(signals, signals, 'full')
        r[:nx] = x[signals.size-1:signals.size+order]
        phi = np.dot(sp.linalg.inv(sp.linalg.toeplitz(r[:-1])), -r[1:])
        return np.concatenate(([1.], phi))
    else:
        return np.ones(1, dtype = signals.dtype)
    
def GETckmatrix(C):
    size =len(C)-0
    Cmatrix = numpy.zeros(size*size).reshape(size,size)
    for x1 in range(0,size):
        for x2 in range(0,size):
            Cmatrix[x1,x2] = C[abs(x2-x1)]
    return Cmatrix

def butterfilter(signal_unfiltered):
    min_len = 60
    if len(signal_unfiltered) < min_len + 1:
        print 'Signal too short'
        return 0
    b, a = signal.butter(8, 0.125)
    list_temp = signal.filtfilt(b, a, signal_unfiltered, padlen=min_len)
    signal_filtered = []
    for x1 in list_temp:
        signal_filtered.append(x1)
    return signal_filtered

def LPC_plot(signal_input, alpha):
    #print 'len(signal_input)',len(signal_input)
    order = len(alpha)
    signal_lpc = []
    signal_diff = []
    for x1 in range(order-1,len(signal_input)):
        ytemp = []
        for x2 in range(0,order):
            value = alpha[x2]*signal_input[x1-x2]
            #print x1-x2, alpha[x2], signal_input[x1-x2]
            ytemp.append(value)
        ynext = sum(ytemp)
        signal_lpc.append(ynext)
        signal_diff.append(ynext-signal_input[x1])
    #print len(signal_input), len(signal_lpc), len(signal_diff)
    figure()
    plt.plot(range(0,len(signal_input)),signal_input, linewidth=1.0, linestyle = '-', color = 'k', label = "Original signal, filtered")
    plt.plot(range(order-1,len(signal_input)),signal_lpc, linewidth=1.2, linestyle = '--',color = 'k', label = "LPC transformed signal")
    plt.plot(range(order-1,len(signal_input)),signal_diff, linewidth=0.7, linestyle = '-',color = 'k', label = "Difference")
    plt.subplots_adjust(right=2.0)
    grid()
    legend(bbox_to_anchor=(1.03, 1), loc=2, borderaxespad=0.)
    return signal_diff, signal_lpc

# Remove excessive data (the oldest)
def data_polisher(feature_speeds):
    x1 = len(feature_speeds) % 24
    if x1 > 0:
        feature_speeds = feature_speeds[x1:]
    return feature_speeds

def LPC_signal_diff(signal_input, alpha): # This is like LPC_plot but with no plot
    order = len(alpha)
    signal_lpc = []
    signal_diff = []
    for x1 in range(order-1,len(signal_input)):
        ytemp = []
        for x2 in range(0,order):
            value = alpha[x2]*signal_input[x1-x2]
            ytemp.append(value)
        ynext = sum(ytemp)
        signal_lpc.append(ynext)
        signal_diff.append(ynext-signal_input[x1])
    return signal_diff, signal_lpc

def GETalpha(C):
    #print 'C',C
    Cmatrix = GETckmatrix(C)
    #print 'len C', len(C)
    #print 'len Cmatrix', len (Cmatrix)
    #print Cmatrix
    Cn = []
    for x1 in range(0,len(C)):
        Cn.append(C[x1]*-1)
    alpha = linalg.solve(Cmatrix[:-1,:-1],Cn[1:])
    #print 'len Cmatrix[:-1,:-1]', len(Cmatrix[:-1,:-1])
    #print 'len alpha',len(alpha)
    #print 'alpha ', alpha
    return alpha

def get_window_sum_plot_absolute(order, window_len, part1_unfiltered, signal_diff, threshold_value):
    #print threshold_value, meanSD(part1_unfiltered)[1],exp(window_len/24)
    signal_len = len(signal_diff)
    threshold_line_u = []
    threshold_line_l = []
    for x1 in range(window_len,signal_len):
        threshold_line_u.append(threshold_value)
    for x1 in range(window_len,signal_len):
        threshold_line_l.append(-threshold_value)
    window_sum = []
    for x1 in range(window_len,signal_len):
        temp = 0
        for x2 in range(0,window_len):
            temp = temp + abs(signal_diff[x1-x2])
        window_sum.append(temp/window_len)
    get_plot(range(order+window_len,order+signal_len), window_sum, "Absolute value of window sum", "LPC-order %s and \n window length %s"%(order, window_len))
    plt.plot(range(order+window_len,order+signal_len), threshold_line_u, color = 'k', linestyle = '--', linewidth = 1.4, label = 'Threshold value')
    return window_sum

def get_window_sum_plot(order, window_len, part1_unfiltered, signal_diff, threshold_value):
    #window_len = order_test+24*6
    signal_len = len(signal_diff)
    #print signal_len-order-window_len, len(signal_diff)
    #threshold_value = meanSD(part1_unfiltered)[1]*2#(5.0+18/exp(window_len/1.7/24))/12
    #print meanSD(part1_unfiltered)[1]*2,exp(window_len/24), threshold_value
    threshold_line_u = []
    threshold_line_l = []
    for x1 in range(window_len,signal_len):
        threshold_line_u.append(threshold_value)
    for x1 in range(window_len,signal_len):
        threshold_line_l.append(-threshold_value)
    window_sum = []
    for x1 in range(window_len,signal_len):
        temp = 0
        for x2 in range(0,window_len):
            temp = temp + signal_diff[x1-x2]
        window_sum.append(temp/window_len)
    #print 'window_sum', len(window_sum)
    #print signal_len-order-window_len, 'fran',order+window_len,'till',signal_len
    get_plot(range(order+window_len,order+signal_len), window_sum, "Window average", "LPC-order %s and \n window length %s"%(order, window_len))
    #print order+window_len, signal_len, len(window_sum)
    plt.plot(range(order+window_len,order+signal_len), threshold_line_u, color = 'k', linestyle = '--', linewidth = 1.4, label = 'Threshold value')
    plt.plot(range(order+window_len,order+signal_len), threshold_line_l, color = 'k', linestyle = '--', linewidth = 1.4)
    return window_sum

def get_window_sum_value_absolute(order, window_len, part1_unfiltered, signal_diff, threshold_value):
    signal_len = len(signal_diff)
    threshold_line_u = []
    threshold_line_l = []
    for x1 in range(window_len,signal_len):
        threshold_line_u.append(threshold_value)
    for x1 in range(window_len,signal_len):
        threshold_line_l.append(-threshold_value)
    window_sum = []
    for x1 in range(window_len,signal_len):
        temp = 0
        for x2 in range(0,window_len):
            temp = temp + abs(signal_diff[x1-x2])
        window_sum.append(temp/window_len)
    return window_sum

def get_window_sum_value(order, window_len, part1_unfiltered, signal_diff, threshold_value):
    signal_len = len(signal_diff)
    threshold_line_u = []
    threshold_line_l = []
    for x1 in range(window_len,signal_len):
        threshold_line_u.append(threshold_value)
    for x1 in range(window_len,signal_len):
        threshold_line_l.append(-threshold_value)
    window_sum = []
    for x1 in range(window_len,signal_len):
        temp = 0
        for x2 in range(0,window_len):
            temp = temp + signal_diff[x1-x2]
        window_sum.append(temp/window_len)
    return window_sum

def GETasymptot_alpha(order_days):
    alpha = []
    for x1 in range(0,order_days): #creating "asymptotical alpha"
        for x1 in range(0,4):
            alpha.append(1)
            alpha.append(0)
            alpha.append(-1)
            alpha.append(-1)
            alpha.append(0)
            alpha.append(1)
        alpha[-1] = 1+1.0/order_days
    alpha.append(0)
    return alpha

def LPC_test(order_days, input_signal,  abs_true_false, window_len_temp, alpha_asympt, plot_true_false):
    order = order_days*24+1
    #threshold_value = 0.35
    #part1_unfiltered = data_polisher(GETfeature_speeds('HTTP')) #create data set and keep at 24*x hr length
    #part2_unfiltered = data_polisher(GETfeature_speeds('Unknown'))
    order = order_days*24+1
    part1_unfiltered = input_signal[:len(input_signal)/2]#-order-24]
    part2_unfiltered = input_signal[len(input_signal)/2:]#+order+24:]
    #print len(part1_unfiltered), len(part2_unfiltered), input_signal
    data_test = []
    original_signal = []
    for x1 in part1_unfiltered:
        original_signal.append(x1)
    for x1 in part2_unfiltered:
        original_signal.append(x1)
    
    part1_filtered = butterfilter(part1_unfiltered) #filter the signal
    part2_filtered = butterfilter(part2_unfiltered)
    signal_filtered = []
    for x1 in part1_filtered:
        signal_filtered.append(x1)
    for x1 in part2_filtered:
        signal_filtered.append(x1)
    
    #print part1_filtered 
    lpc_coeff = lpc_ref(part1_filtered, order) #receive LPC coeff from the first part
    #print 'lpc_coeff', sum(lpc_coeff)#, lpc_coeff
    if alpha_asympt == 0:
        alpha = GETalpha(lpc_coeff) #use LPC coeff to obtain alpha   #-lpc_coeff[1:]#
    elif alpha_asympt == 1:
        alpha = GETasymptot_alpha(order_days)
    else:
        print 'alpha_asympt must be equal to zero or one'
    #print 'alpha', sum(alpha), alpha
    alphasum = sum(alpha)
    lpc_coeff = []
    for x1 in range(0,len(alpha)):
        lpc_coeff.append(alpha[x1] / alphasum) #normalize alpha
    signal_test = []
    for x1 in part1_unfiltered: #create a test signal (unfiltered)
        signal_test.append(x1)
    for x1 in part2_unfiltered:
        signal_test.append(x1)
    
    #figure()
    #plt.plot(signal_test, color = 'k', label = "Input signal, \n unfiltered")
    #plt.subplots_adjust(right=2.0)
    #legend(bbox_to_anchor=(1.03, 1), loc=2, borderaxespad=0.)
    signal_test = butterfilter(signal_test) #then filter the signal
    if plot_true_false == 0:
        signal_diff = LPC_signal_diff(signal_test, lpc_coeff)[0]
    elif plot_true_false == 1 :
        signal_diff = LPC_plot(signal_test, lpc_coeff)[0]
    else:
        print 'plot_true_false must be equal to zero or one'
    #signal_lpc = LPC_plot(signal_test, lpc_coeff)[1]
    
    #print '************************************************* OUTCOME ********'
    if abs_true_false == 0:
        threshold_value = (meanSD(part1_unfiltered)[0]+meanSD(part1_unfiltered)[1])*0.03*(1+1.0/order_days)*(1+168.0/window_len_temp)
        #print 'coeff', (2-1.0/order_days)*(1+48.0/window_len_temp), (2-1.0/order_days), (1+48.0/window_len_temp)
        #print 'threshold', threshold_value
        if plot_true_false == 0:
            window_sum_regular = get_window_sum_value(order, window_len_temp, part1_unfiltered, signal_diff, threshold_value)
        elif plot_true_false == 1 :
            window_sum_regular = get_window_sum_plot(order, window_len_temp, part1_unfiltered, signal_diff, threshold_value)
        #print 'Normal for window length', window_len_temp,' window sum',max(max(window_sum_regular[:336-order-window_len_temp])/1000000, min(window_sum_regular)/1000000)
        #print 'Anomaly for window length', window_len_temp,' window sum:', max(max(window_sum_regular[336-order-window_len_temp:])/1000000, abs(min(window_sum_regular[336-order-window_len_temp:])/1000000))
        part1_window_max = max(max(window_sum_regular[:336-order-window_len_temp]), abs(min(window_sum_regular[:336-order-window_len_temp])))
        part2_window_max = max(max(window_sum_regular[336-order-window_len_temp:]), abs(min(window_sum_regular[336-order-window_len_temp:])))
        #print 'regular th', threshold_value, 'part 1 and 2 max', part1_window_max, part2_window_max 
    elif abs_true_false == 1:
        threshold_value = (meanSD(part1_unfiltered)[0]+meanSD(part1_unfiltered)[1])*0.2*(1+1.0/order_days)*(1+24.0/window_len_temp) #KOLLA PÅ VÄRDEN UTAN COEFF OCH SEDAN GÖR EN RIKTIG FORMEL SOM STÄMMER FÖR DE VÄRDENA
        #window_len_temp = order_days*24 # For detecting local anomalies
        if plot_true_false == 0:
            window_sum_abs = get_window_sum_value_absolute(order, window_len_temp, part1_unfiltered, signal_diff, threshold_value)
        elif plot_true_false == 1 :
            window_sum_abs= get_window_sum_plot_absolute(order, window_len_temp, part1_unfiltered, signal_diff, threshold_value)
        #print 'Normal for window length', window_len_temp,' absolute window sum',max(window_sum_abs[:336-order-window_len_temp])/1000000
        #print 'Anomaly for window length', window_len_temp,' absolute',max(window_sum_abs[336-order-window_len_temp:])/1000000
        part1_window_max = max(window_sum_abs[:336-order-window_len_temp])
        part2_window_max = max(window_sum_abs[336-order-window_len_temp:])
        #print 'abs th', threshold_value, 'part 1 and 2 max', part1_window_max, part2_window_max
    else:
        print 'abs_true_false must be 1 or 0'
    #pyplot.close()
    
    part1_anomaly_detected = part1_window_max/threshold_value
    part2_anomaly_detected = part2_window_max/threshold_value
    #part1_anomaly_detected = 0
    #if part1_window_max > threshold_value:
    #    part1_anomaly_detected = 1
    #part2_anomaly_detected = 0
    #if part2_window_max > threshold_value:
    #    part2_anomaly_detected = 1
    
    return round(part1_anomaly_detected,2), round(part2_anomaly_detected,2)

def get_plot(time, signal, title_text, label_text):
    figure()
    title(title_text)
    ylabel('Signal [bytes/s]')
    xlabel('Time [h]')
    plt.subplots_adjust(right=1.6)
    grid()
    if time == 0:
        if label_text == 0:
            plt.plot(signal, color = 'k')
        else:
            plt.plot(signal, color = 'k', label = label_text)
            legend(bbox_to_anchor=(1.03, 1), loc=2, borderaxespad=0.)
            
    else:
        if label_text == 0:
            plt.plot(time, signal, color = 'k')
        else:
            plt.plot(time, signal, color = 'k', label = label_text)
            legend(bbox_to_anchor=(1.03, 1), loc=2, borderaxespad=0.)
    return
```
