```python
import numpy as np
import scipy as sp
import scipy.signal as sig
#import scikits
#from scikits.talkbox.linpred.lpc import lpc

def lpc_ref(signal, order):
    """Compute the Linear Prediction Coefficients.

    Return the order + 1 LPC coefficients for the signal. c = lpc(x, k) will
    find the k+1 coefficients of a k order linear filter:

      xp[n] = -c[1] * x[n-2] - ... - c[k-1] * x[n-k-1]

    Such as the sum of the squared-error e[i] = xp[i] - x[i] is minimized.

    Parameters
    ----------
    signal: array_like
        input signal
    order : int
        LPC order (the output will have order + 1 items)

    Notes
    ----
    This is just for reference, as it is using the direct inversion of the
    toeplitz matrix, which is really slow"""
    if isinstance(signal, list):
        signal = np.asarray(signal)
    if signal.ndim > 1:
        raise ValueError("Array of rank > 1 not supported yet")
    if order > len(signal):
        raise ValueError("Input signal must have a lenght >= lpc order")

    if order > 0:
        p = order + 1
        r = np.zeros(p, signal.dtype)
        # Number of non zero values in autocorrelation one needs for p LPC
        # coefficients
        nx = np.min([p, signal.size])
        x = np.correlate(signal, signal, 'full')
        r[:nx] = x[signal.size-1:signal.size+order]
        phi = np.dot(sp.linalg.inv(sp.linalg.toeplitz(r[:-1])), -r[1:])
        return np.concatenate(([1.], phi))
    else:
        return np.ones(1, dtype = signal.dtype)
    
def GETckmatrix(C):
    size =len(C)-0
    Cmatrix = numpy.zeros(size*size).reshape(size,size)
    for x1 in range(0,size):
        for x2 in range(0,size):
            Cmatrix[x1,x2] = C[abs(x2-x1)]
    return Cmatrix

def butterfilter(signal_unfiltered):
    min_len = 60
    if len(signal_unfiltered) < min_len + 1:
        print 'Signal too short'
        return 0
    b, a = signal.butter(8, 0.125)
    list_temp = signal.filtfilt(b, a, signal_unfiltered, padlen=min_len)
    signal_filtered = []
    for x1 in list_temp:
        signal_filtered.append(x1)
    return signal_filtered

def meanSDdetrend(signal_l):
    signal_out = []
    f1 = LSM(signal_l)[0]
    for x1 in range(0,len(signal_l)):
        temp1 = signal_l[x1]
        temp2 = f1(x1)
        temp3 = meanSD(signal_l)[1]
        temp4 = temp1-temp2
        temp5 = temp4/temp3
        signal_out.append(temp5)
    return signal_out
        
old_data = GETfeature_speeds('Unknown')[0:]
#extraa = GETfeature_speeds('HTTP')
#for x1 in extraa:
#    old_data.append(x1)
#print len(old_data)
#mylistt = old_data


signal_l = butterfilter(old_data)

#print 'haeeeeaeaeaeaaeeaer', mylist
##orderV = [3, 5, 9, 15, 24, 48, 72, 96]
def LPC_plot(signal_l, orderV,):
    x_figure = 0
    for order in orderV:
        C = lpc_ref(signal_l[:], order)
        #C = [1, -0.768, -0.3299, 0.1474]
        #C = lpc.kautocor(mylist, order)
        #print 'C',C
        ##Cmatrix = GETckmatrix(C)
        #print Cmatrix
        ##Cflip = []
        Cn = []
        for x1 in range(0,len(C)):
            ##Cflip.append(C[-x1])
            Cn.append(C[x1]*-1)
        ##alpha = linalg.solve(Cmatrix[:-1,:-1],Cn[1:])
        ##alpha_original = linalg.solve(Cmatrix[:-1,:-1],C[1:]) #fel
        #print 'alpha', alpha
        #alpha0 = alpha[0]
        #print 'alpha0 = ', alpha0
        #for x4 in range(0,len(alpha)):
        #    alpha[x4] = alpha[x4] / alpha0
        #    alpha_original[x4] = alpha_original[x4] / alpha0
        #    print 'alpha nummer ', x4, ' aer lika med ', alpha[x4]
        #print 'alpha', alpha, alpha_original
        #print 'Cn', Cn[1:]
        #print 'Cmatricx'
        #print Cmatrix[:-1,:-1]
        signal_lpc = []
        signal_diff = []
        for x1 in range(0,len(signal_l)-0):
            ytemp = []
            #alpha = linalg.solve(Cmatrix[:-1,:-1],Cn[1:])
            #alpha0 = alpha[0]
            #print 'alpha0 = ', alpha0
            #for x4 in range(0,len(alpha)):
            #    alpha[x4] = alpha[x4] / alpha0
                #print 'alpha nummer ', x4, ' aer lika med ', alpha[x4]
            #print 'alpha', alpha, alpha_original
            #print 'originalalpha', alpha, alpha_original
            #for x3 in range(0,order):
            #    if x1 < x3:
            #        alpha[x3] = 0
            
            #print 'omgang', x1, ' med alpha = ', alpha, ' och mylist = ', mylist[0:10]
            #if x1 < order-1:
            #    mylist2.append(mylist[x1])
            #else:
            for x2 in range(0,len(C)-1):
                if -(x2+0)+x1 < 0:
                    varde2 = 0
                else:
                    varde = C[x2+1] * signal_l[0-(x2+1)+x1]
                    varde2 = varde * -1
                #if x1 < 10:
                #    print 'varde2 dvs det under blir', varde2
                #    print C[x2+1], 'gånger', mylist[-(x2+0)+x1], -(x2+0)+x1
                #    print mylist[0:6]
                #print 'varde2', varde2
                ytemp.append(varde2)
                #print 'ytemp', ytemp
            ynext = sum(ytemp)
            if x1 < 2.5:
                ynext = signal_l[x1]
            #print 'summan blev ', ynext
            signal_lpc.append(ynext)
            signal_diff.append(0-signal_lpc[x1]+signal_l[x1])
        x_figure = x_figure + 1
        figure(x_figure)
        plt.plot(range(0,len(sisignal_lgnal)),signal_l[:], label = "original signal")
        plt.plot(range(0,len(signal_lpc)),signal_lpc, label = "LPC transformed signal")
        #figure(3)
        plt.plot(range(0,len(signal_diff)),signal_diff, label = "difference")
        grid()
        legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        print 'Order', order, 'has norm', linalg.norm(signal_diff)
        return 1

def LPC_plot2(signal_l, C, signal_name, figure_number):
    x_figure = 0
    order = len(C)-1
    ##C = lpc_ref(signal[:], order)
    #C = [1, -0.768, -0.3299, 0.1474]
    #C = lpc.kautocor(mylist, order)
    #print 'C',C
    ##Cmatrix = GETckmatrix(C)
    #print Cmatrix
    ##Cflip = []
    Cn = []
    for x1 in range(0,len(C)):
        ##Cflip.append(C[-x1])
        Cn.append(C[x1]*-1)
    ##alpha = linalg.solve(Cmatrix[:-1,:-1],Cn[1:])
    ##alpha_original = linalg.solve(Cmatrix[:-1,:-1],C[1:]) #fel
    #print 'alpha', alpha
    #alpha0 = alpha[0]
    #print 'alpha0 = ', alpha0
    #for x4 in range(0,len(alpha)):
    #    alpha[x4] = alpha[x4] / alpha0
    #    alpha_original[x4] = alpha_original[x4] / alpha0
    #    print 'alpha nummer ', x4, ' aer lika med ', alpha[x4]
    #print 'alpha', alpha, alpha_original
    #print 'Cn', Cn[1:]
    #print 'Cmatricx'
    #print Cmatrix[:-1,:-1]
    signal_lpc = []
    signal_diff = []
    for x1 in range(0,len(signal_l)-0):
        ytemp = []
        #alpha = linalg.solve(Cmatrix[:-1,:-1],Cn[1:])
        #alpha0 = alpha[0]
        #print 'alpha0 = ', alpha0
        #for x4 in range(0,len(alpha)):
        #    alpha[x4] = alpha[x4] / alpha0
            #print 'alpha nummer ', x4, ' aer lika med ', alpha[x4]
        #print 'alpha', alpha, alpha_original
        #print 'originalalpha', alpha, alpha_original
        #for x3 in range(0,order):
        #    if x1 < x3:
        #        alpha[x3] = 0
        
        #print 'omgang', x1, ' med alpha = ', alpha, ' och mylist = ', mylist[0:10]
        #if x1 < order-1:
        #    mylist2.append(mylist[x1])
        #else:
        for x2 in range(0,len(C)-0):
            if -(x2+0)+x1 < 0:
                varde2 = 0
            else:
                varde = C[x2+0] * signal_l[0-(x2+0)+x1]
                varde2 = varde * -1
            #if x1 < 10:
            #    print 'varde2 dvs det under blir', varde2
            #    print C[x2+1], 'gånger', mylist[-(x2+0)+x1], -(x2+0)+x1
            #    print mylist[0:6]
            #print 'varde2', varde2
            ytemp.append(varde2)
            #print 'ytemp', ytemp
        ynext = sum(ytemp)
        if x1 < 2.5: # ignorera tre första resultat :S
            ynext = abs(signal_l[x1])
        #print 'summan blev ', ynext
        signal_lpc.append(abs(ynext))
        signal_diff.append(signal_lpc[x1]-signal_l[x1])
    x_figure = x_figure + 1
    figure(figure_number)
    plt.plot(range(0,len(signal_l)),signal_l[:], label = "original signal")
    plt.plot(range(0,len(signal_lpc)),signal_lpc, label = "LPC transformed signal")
    #figure(3)
    plt.plot(range(0,len(signal_diff)),signal_diff, label = "difference")
    grid()
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    print 'Service:',signal_name, 'with order', order, 'has norm', linalg.norm(signal_diff), 'and meanSD =', meanSD(signal_diff)
    return signal_diff

def GETalpha(C):
    print 'C',C
    Cmatrix = GETckmatrix(C)
    print 'len C', len(C)
    print 'len Cmatrix', len (Cmatrix)
    #print Cmatrix
    Cn = []
    for x1 in range(0,len(C)):
        Cn.append(C[x1]*-1)
    alpha = linalg.solve(Cmatrix[:-1,:-1],Cn[1:])
    print 'len Cmatrix[:-1,:-1]', len(Cmatrix[:-1,:-1])
    print 'len alpha',len(alpha)
    print 'alpha ', alpha
    return alpha

def LPC_plot2_procent(signal_l, C, signal_name, figure_number):
    x_figure = 0
    order = len(C)-0
    signal_lpc = []
    signal_diff = []
    for x1 in range(0,len(signal_l)-0):
        ytemp = []
        for x2 in range(0,order):
            if -(x2+0)+x1 < 0:
                varde2 = 0
            else:
                varde = C[x2+0] * signal_l[0-(x2+1)+x1]
                varde2 = varde * -1
            ytemp.append(varde2)
        ynext = sum(ytemp)
        if x1 < 2.5:
            ynext = abs(signal_l[x1])
        #print 'summan blev ', ynext
        signal_lpc.append(abs(ynext))
        if (signal_lpc[x1]+signal_l[x1]) == 0:
            diffproc = 0
        else:
            diffproc = (signal_l[x1] - signal_lpc[x1]) / (signal_lpc[x1]+signal_l[x1])
        signal_diff.append(diffproc)
    x_figure = x_figure + 1
    figure(figure_number)
    plt.plot(range(0,len(signal_l)),signal_l[:], label = "original signal")
    plt.plot(range(0,len(signal_lpc)),signal_lpc, label = "LPC transformed signal")
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    figure(figure_number+1)
    plt.plot(range(0,len(signal_diff)),signal_diff, label = "difference [procent]")
    grid()
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    print 'Service:',signal_name, 'with order', order, 'has norm', linalg.norm(signal_diff), 'and meanSD =', meanSD(signal_diff)
    return signal_diff

def MdiagADD(quantity, size):
    if (quantity+1)%2==0:
        print "good"
    else:
        print "Please enter odd integers"
        return 0
    y = numpy.zeros(size**2).reshape((size,size))
    v = []
    for x1 in range(0,quantity):
        v.append(1)
    r = -(quantity - 1) / 2
    for x1 in v:
        s = []
        for x2 in range(0,size-abs(r)):
            s.append(x1)
        y2 = np.diag(s,r)
        y = y + y2
        r = r + 1
    for x1 in range(0,len(y)):
        x3 = sum(y[x1])
        for x2 in range(0,len(y)):
            y[x1,x2] = y[x1,x2] / x3
    return y
```
