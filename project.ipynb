```python
def get_time(r):
    return r['systemtime']

def meanSD(x):
    mean = sum(x)/len(x) #mean value
    m=[]
    for y in x:
        k = (y-mean)**2
        m.append(k)
    SD = sqrt(sum(m)/len(x)) #standard deviation
    return mean, SD

def getSpeedForService(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['in_speed']+x['out_speed']
    return 0

def error(f, x, y):
    return sp.sum((f(x)-y)**2)

def collect_services(reportList):
    resultatet = set()
    for x in reportList:
        for y in x['services']:
            resultatet.add(y['name'])
    return resultatet
```
