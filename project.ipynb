```python
def getSpeedForService(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['in_speed']+x['out_speed']
    return 0
```

```python
http_speeds = []
for x in reports:
    http_speeds.append(getSpeedForService(x, 'HTTP'))
http_speeds
```

```python
pylab.plot(reptimes, http_speeds)
```

```python
def collect_services(reportList):
    resultatet = set()
    for x in reportList:
        for y in x['services']:
            resultatet.add(y['name'])
    return resultatet
```

```python
collect_services(reports)
```

```python
data2 = []
for servicename in collect_services(reports):
    speeds = []
    speeds.append(servicename)
    for x in reports[:]:
        speeds.append(getSpeedForService(x, servicename))
    data2.append(speeds)
data2
```
