```python
def getSpeedForService(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['in_speed']+x['out_speed']
    return 0
```

```python
http_speeds = []
for x in reports:
    http_speeds.append(getSpeedForService(x, 'HTTP'))
http_speeds
```

```python
def collect_services(reportList):
    resultatet = set()
    for x in reportList:
        for y in x['services']:
            resultatet.add(y['name'])
    return resultatet
```
