```python
import json, pprint, pylab, numpy, scipy, math
import matplotlib.pyplot as plt
import scipy as sp

f = file('C:\\Users\\arnarn\\desktop\\services-for-exjobb2.json')
data = json.load(f)
data.keys()
reports = data['reports']
tid1 = reports[0]['reporttime']
tid2 = reports[1]['reporttime']
reptimes = map(get_time, reports)
```

```python
def get_time(r):
    return r['systemtime']

def meanSD(x):
    mean = sum(x)/len(x) #mean value
    m=[]
    for y in x:
        k = (y-mean)**2
        m.append(k)
    SD = sqrt(sum(m)/len(x)) #standard deviation
    return mean, SD

def getSpeedForService(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['in_speed']+x['out_speed']
    return 0

def error(f, x, y):
    return sp.sum((f(x)-y)**2)

def collect_services(reportList):
    resultatet = set()
    for x in reportList:
        for y in x['services']:
            resultatet.add(y['name'])
    return resultatet

def GETfeature_speeds(servicename):
    feature_speeds = []
    for x in reports:
        feature_speeds.append(getSpeedForService(x, servicename))
    return feature_speeds

# Rework to 24 daily reports (a certain time for every day)
def GETmatris(feature_speeds):
    feature_speeds_daily = []
    started_days = (len(feature_speeds)/24) + (len(feature_speeds) % 24 > 0)
    hour = range(0,24)
    matris = numpy.zeros(len(hour)*started_days).reshape((started_days,len(hour)))
    y1=-1
    y2=-1
    for x1 in hour:
        days = range(x1,len(feature_speeds),24)
        for x2 in days:
            feature_speeds_daily.append(feature_speeds[x2])
    for x1 in hour:
        y1=y1+1
        y2=-1
        days = range(x1,len(feature_speeds),24)
        for x2 in days:
            y2=y2+1
            matris[y2,y1]=feature_speeds[x2]
    return matris

def GETvariables(matris, feature_speeds): # unused_hours, started_days, finished_days
    hour = range(0,24)
    unused_hours = 0
    for x1 in hour:
        days = range(x1,len(feature_speeds),24)
        speeds_for_hour = matris[:,x1]
        if len(days) < len(matris[:,x1]):
            unused_hours = unused_hours+1
    started_days = (len(feature_speeds)/24) + (len(feature_speeds) % 24 > 0)
    finished_days = (len(feature_speeds)/24)
    return unused_hours, started_days, finished_days

# Create SDtotMean value, i.e. mean value of every diff for all data.
def GETSDtotMean(matris, feature_speeds):
    hour = range(0,24)
    SDtotList = []
    for x1 in hour:
        days = range(x1,len(feature_speeds),24)
        speeds_for_hour = matris[:,x1]
        if len(days) < len(matris[:,x1]):
            speeds_for_hour = matris[0:-1,x1]  # Remove the last zero digit if the full 24 hr of the day hasn't passed
        diff_for_hour = [j-i for i, j in zip(speeds_for_hour[:-1], speeds_for_hour[1:])]
        SDtotElement = meanSD(speeds_for_hour)[1]
        SDtotList.append(SDtotElement)
    SDtotMean = meanSD(SDtotList)[0]
    return SDtotMean

def GETlist_diff(list1,list2):
    list_diff = []
    for x1 in range(0,len(list1)):
        diff = list2[x1] - list1[x1]
        list_diff.append(diff)
    return list_diff
```

```python
# Moving window
def GETMax_Average(Data, Average_Of):
    temp = Data
    temp.sort()
    sum_maxV = []
    for x1 in range(0,Average_Of):
        sum_maxV.append(temp[-x1-1])
    average_maxV = sum(sum_maxV) / Average_Of
    return average_maxV

def GETMin_Average(Data, Average_Of):
    temp = Data
    temp.sort()
    sum_minV = []
    for x1 in range(0,Average_Of):
        sum_minV.append(temp[x1])
    average_minV = sum(sum_minV) / Average_Of
    return average_minV

def GETline_fit(list_average, days):
    # Fitted line for max value
    startV = (len(list_average[:-1])-days)*24+12
    endV = (len(list_average)-1)*24+12
    fp1, residuals, rank, sv, rcond = sp.polyfit(range(startV,endV,24),list_average[-days-1:-1],1,full=True)
    f1=sp.poly1d(fp1)
    fx = sp.linspace(startV,endV, 30) 
    return f1, fx



def Moving_Window(Using_Days, Using_Average_Of, Using_Feature_Speeds):
    
    average_of = Using_Average_Of
    print servicename
    feature_speeds = Using_Feature_Speeds
    matris = GETmatris(feature_speeds)
    variables = GETvariables(matris, feature_speeds)
    unused_hours = variables[0]
    if unused_hours == 0:
        started_days = variables[1]
    else:
        feature_speeds = feature_speeds[(24-unused_hours):]
        matris = GETmatris(feature_speeds)
        variables = GETvariables(matris, feature_speeds)
        unused_hours = variables[0]
        started_days = variables[1]
    average_maxV = [] 
    average_minV = [] 
    plt.plot(range(0,len(feature_speeds)),feature_speeds)
    list_average_maxV = []
    list_average_minV = []
    for x1 in range(0,started_days):
        average_maxV = GETMax_Average(matris[x1],average_of)
        average_minV = GETMin_Average(matris[x1],average_of)
        list_average_maxV.append(average_maxV)
        list_average_minV.append(average_minV)
    
    plot_average_maxV = []
    plot_average_minV = []
    for x2 in range(0,len(list_average_maxV)):
        for x3 in range(0,24):
            plot_average_maxV.append(list_average_maxV[x2])
            plot_average_minV.append(list_average_minV[x2])
    plt.plot(range(0,len(feature_speeds)),plot_average_maxV, range(0,len(feature_speeds)),plot_average_minV)
    Max_Line = GETline_fit(list_average_maxV, Using_Days)
    Min_Line = GETline_fit(list_average_minV, Using_Days)
    fx = Max_Line[1]
    f1 = Max_Line[0]
    f2 = Min_Line[0]
    plt.plot(fx, f1(fx), linewidth=4)
    plt.plot(fx, f2(fx), linewidth=4)
    return 0 #len(matris), len(feature_speeds), started_days, list_average_maxV
```
