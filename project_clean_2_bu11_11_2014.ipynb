```python


# 'out_cps', 'in', 'conns', 'in_cps', 'out', 'out_speed', 'in_speed'

#def GETtot_data():
#    tot_data = []
#    for x3 in range(0,len(GETfeature_speeds('HTTP'))):
#        tot_data.append(0)
#        
#    all_reports = collect_services(reports)
#    for x1 in all_reports:
#        temp_feature_speed = GETfeature_speeds(x1)
#        for x2 in range(0,len(temp_feature_speed)):
#            tot_data[x2] = tot_data[x2] + temp_feature_speed[x2]
#    return tot_data

def get_time(r):
    return r['systemtime']

def meanSD(x):
    temp = 0
    for x1 in x:
        temp = temp + x1
    mean = temp/len(x) #mean value
    m=[]
    for y in x:
        k = (y-mean)**2
        m.append(k)
    temp2 = 0
    for x2 in m:
        temp2 = temp2 + x2
    SD = sqrt(temp2/len(x)) #standard deviation
    return mean, SD

def getSpeedForService(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['in_speed']+x['out_speed']
    return 0

def GETout_speed(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['out_speed']
    return 0

def GETin_speed(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['in_speed']
    return 0

def GETin_cpu(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['conns']
    return 0

def GETout_cpu(report, sName):
    for x in report['services']:
        if sName == x['name']:
            return x['out']
    return 0

def error(f, x, y):
    return sp.sum((f(x)-y)**2)

def collect_services(reportList):
    result = set()
    for x in reportList:
        for y in x['services']:
            result.add(y['name'])
    return result

def GETfeature_out_speeds(servicename):
    feature_out_speeds = []
    for x in reports:
        feature_out_speeds.append(GETout_speed(x, servicename))
    return feature_out_speeds

def GETfeature_in_speeds(servicename):
    feature_in_speeds = []
    for x in reports:
        feature_in_speeds.append(GETin_speed(x, servicename))
    return feature_in_speeds

def GETfeature_in_cpu(servicename):
    feature_in_cpu = []
    for x in reports:
        feature_in_cpu.append(GETin_cpu(x, servicename))
    return feature_in_cpu

def GETfeature_out_cpu(servicename):
    feature_out_cpu = []
    for x in reports:
        feature_out_cpu.append(GETout_cpu(x, servicename))
    return feature_out_cpu

def GETfeature_speeds(servicename):
    feature_speeds = []
    for x in reports:
        feature_speeds.append(getSpeedForService(x, servicename))
    return feature_speeds

# Remove excessive data (the oldest)
def data_polisher(feature_speeds):
    x1 = len(feature_speeds) % 24
    if x1 > 0:
        feature_speeds = feature_speeds[x1:]
    return feature_speeds
        

# Rework to 24 daily reports (a certain time for every day)
def GETmatris(feature_speeds):
    feature_speeds_daily = []
    started_days = (len(feature_speeds)/24) + (len(feature_speeds) % 24 > 0)
    hour = range(0,24)
    matris = numpy.zeros(len(hour)*started_days).reshape((started_days,len(hour)))
    y1=-1
    y2=-1
    for x1 in hour:
        days = range(x1,len(feature_speeds),24)
        for x2 in days:
            feature_speeds_daily.append(feature_speeds[x2])
    for x1 in hour:
        y1=y1+1
        y2=-1
        days = range(x1,len(feature_speeds),24)
        for x2 in days:
            y2=y2+1
            matris[y2,y1]=feature_speeds[x2]
    return matris

def GETvariables(matris, feature_speeds): # unused_hours, started_days, finished_days
    hour = range(0,24)
    unused_hours = 0
    for x1 in hour:
        days = range(x1,len(feature_speeds),24)
        speeds_for_hour = matris[:,x1]
        if len(days) < len(matris[:,x1]):
            unused_hours = unused_hours+1
    started_days = (len(feature_speeds)/24) + (len(feature_speeds) % 24 > 0)
    finished_days = (len(feature_speeds)/24)
    return unused_hours, started_days, finished_days

# Create SDtotMean value, i.e. mean value of every diff for all data.
def GETSDtotMean(matris, feature_speeds):
    hour = range(0,24)
    SDtotList = []
    for x1 in hour:
        days = range(x1,len(feature_speeds),24)
        speeds_for_hour = matris[:,x1]
        if len(days) < len(matris[:,x1]):
            speeds_for_hour = matris[0:-1,x1]  # Remove the last zero digit if the full 24 hr of the day hasn't passed
        diff_for_hour = [j-i for i, j in zip(speeds_for_hour[:-1], speeds_for_hour[1:])]
        SDtotElement = meanSD(speeds_for_hour)[1]
        SDtotList.append(SDtotElement)
    SDtotMean = meanSD(SDtotList)[0]
    return SDtotMean

def GETlist_diff(list1,list2):
    list_diff = []
    for x1 in range(0,len(list1)):
        diff = list2[x1] - list1[x1]
        list_diff.append(diff)
    return list_diff

# Creates a treshold for the plot given as input y
def SDplot(k, y, reference_length):
    if reference_length == 0:
        reference_length = len(y)
    SD = meanSD(y[0:reference_length])[1]
    line = []
    nline = []
    for x1 in y:
        line.append(SD*k)
        nline.append(SD*-1*k)
    plt.plot(range(0,len(y)),line)
    plt.plot(range(0,len(y)),nline)
    return SD

# Feature traffic over total traffic
#def feature_over_total(feature_data):
#    fraction = []
#    if len(feature_data) > len(tot_data):
#        feature_data = feature_data[-len(tot_data):]
#        print 'feature data has been shortened'
#    elif len(feature_data) < len(tot_data):
#        return [0,'feature data too short']
#    for x1 in range(0,len(feature_data)):
#        frac = feature_data[x1] / float(tot_data[x1])
#        fraction.append(frac)
#    #plt.plot(range(0,len(fraction)),fraction)
#    return fraction

# Creates a line (linear eq.) through the plot using Least Square Method
def LSM(y):
    x = range(0,len(y))
    fp1, residuals, rank, sv, rcond = sp.polyfit(x,y,1,full=True)
    #print("model parameters: %s" % fp1)
    f1=sp.poly1d(fp1)
    fx = sp.linspace(0,x[-1],1000) 
    return f1, fx

def split_data(data, new_points):
    reference_data = data[0:-new_points]
    new_data = data[-new_points:]
    return reference_data, new_data

# Moving window
def GETMax_Average(Data, Average_Of):
    temp = Data
    temp.sort()
    sum_maxV = []
    for x1 in range(0,Average_Of):
        sum_maxV.append(temp[-x1-1])
    average_maxV = sum(sum_maxV) / Average_Of
    return average_maxV

def GETMin_Average(Data, Average_Of):
    temp = Data
    temp.sort()
    sum_minV = []
    for x1 in range(0,Average_Of):
        sum_minV.append(temp[x1])
    average_minV = sum(sum_minV) / Average_Of
    return average_minV

def GETline_fit(list_average, days):
    # Fitted line for max value
    startV = (len(list_average[:-1])-days)*24+12
    endV = (len(list_average)-1)*24+12
    fp1, residuals, rank, sv, rcond = sp.polyfit(range(startV,endV,24),list_average[-days-1:-1],1,full=True)
    f1=sp.poly1d(fp1)
    fx = sp.linspace(startV,endV, 30) 
    return f1, fx



def Moving_Window(Using_Days, Using_Average_Of, Using_Feature_Speeds):
    average_of = Using_Average_Of
    print servicename
    feature_speeds = Using_Feature_Speeds
    matris = GETmatris(feature_speeds)
    variables = GETvariables(matris, feature_speeds)
    unused_hours = variables[0]
    if unused_hours == 0:
        started_days = variables[1]
    else:
        feature_speeds = feature_speeds[(24-unused_hours):]
        matris = GETmatris(feature_speeds)
        variables = GETvariables(matris, feature_speeds)
        unused_hours = variables[0]
        started_days = variables[1]
    average_maxV = [] 
    average_minV = [] 
    plt.plot(range(0,len(feature_speeds)),feature_speeds)
    list_average_maxV = []
    list_average_minV = []
    for x1 in range(0,started_days):
        average_maxV = GETMax_Average(matris[x1],average_of)
        average_minV = GETMin_Average(matris[x1],average_of)
        list_average_maxV.append(average_maxV)
        list_average_minV.append(average_minV)
    
    plot_average_maxV = []
    plot_average_minV = []
    for x2 in range(0,len(list_average_maxV)):
        for x3 in range(0,24):
            plot_average_maxV.append(list_average_maxV[x2])
            plot_average_minV.append(list_average_minV[x2])
    plt.plot(range(0,len(feature_speeds)),plot_average_maxV, range(0,len(feature_speeds)),plot_average_minV)
    Max_Line = GETline_fit(list_average_maxV, Using_Days)
    Min_Line = GETline_fit(list_average_minV, Using_Days)
    fx = Max_Line[1]
    f1 = Max_Line[0]
    f2 = Min_Line[0]
    plt.plot(fx, f1(fx), linewidth=4)
    plt.plot(fx, f2(fx), linewidth=4)
    plt.grid()
    max_diff = list_average_maxV[-1] - f1(fx[-1])
    min_diff = list_average_minV[-1] - f2(fx[-1])
    max_SD = meanSD(list_average_maxV[:-1])[1]
    min_SD = meanSD(list_average_minV[:-1])[1]
    return max_diff, min_diff, max_SD, min_SD #len(matris), len(feature_speeds), started_days, list_average_maxV

def signal_damaged():
    data_unknown = data_polisher(GETfeature_speeds('Unknown'))
    data_http = data_polisher(GETfeature_speeds('HTTP'))
    signal_unfiltered = []
    for x1 in data_unknown:
        signal_unfiltered.append(x1)
    for x1 in data_http:
        signal_unfiltered.append(x1)
    return signal_unfiltered

# Constant variables
reptimes = map(get_time, reports)
#tot_data = GETtot_data()
HTTP_data = GETfeature_speeds('HTTP')

import numpy as np
import scipy as sp
import scipy.signal as sig

def lpc_ref(signal, order):
    """Compute the Linear Prediction Coefficients.

    Return the order + 1 LPC coefficients for the signal. c = lpc(x, k) will
    find the k+1 coefficients of a k order linear filter:

      xp[n] = -c[1] * x[n-2] - ... - c[k-1] * x[n-k-1]

    Such as the sum of the squared-error e[i] = xp[i] - x[i] is minimized.

    Parameters
    ----------
    signal: array_like
        input signal
    order : int
        LPC order (the output will have order + 1 items)

    Notes
    ----
    This is just for reference, as it is using the direct inversion of the
    toeplitz matrix, which is really slow"""
    if isinstance(signal, list):
        signal = np.asarray(signal)
    if signal.ndim > 1:
        raise ValueError("Array of rank > 1 not supported yet")
    if order > len(signal):
        raise ValueError("Input signal must have a lenght >= lpc order")

    if order > 0:
        p = order + 1
        r = np.zeros(p, signal.dtype)
        # Number of non zero values in autocorrelation one needs for p LPC
        # coefficients
        nx = np.min([p, signal.size])
        x = np.correlate(signal, signal, 'full')
        r[:nx] = x[signal.size-1:signal.size+order]
        phi = np.dot(sp.linalg.inv(sp.linalg.toeplitz(r[:-1])), -r[1:])
        return np.concatenate(([1.], phi))
    else:
        return np.ones(1, dtype = signal.dtype)
    
def GETckmatrix(C):
    size =len(C)-0
    Cmatrix = numpy.zeros(size*size).reshape(size,size)
    for x1 in range(0,size):
        for x2 in range(0,size):
            Cmatrix[x1,x2] = C[abs(x2-x1)]
    return Cmatrix

def butterfilter(signal_unfiltered):
    min_len = 60
    if len(signal_unfiltered) < min_len + 1:
        print 'Signal too short'
        return 0
    b, a = signal.butter(8, 0.125)
    list_temp = signal.filtfilt(b, a, signal_unfiltered, padlen=min_len)
    signal_filtered = []
    for x1 in list_temp:
        signal_filtered.append(x1)
    return signal_filtered

def meanSDdetrend(signal_l):
    signal_out = []
    f1 = LSM(signal_l)[0]
    for x1 in range(0,len(signal_l)):
        temp1 = signal_l[x1]
        temp2 = f1(x1)
        temp3 = meanSD(signal_l)[1]
        temp4 = temp1-temp2
        temp5 = temp4/temp3
        signal_out.append(temp5)
    return signal_out
        
#old_data = GETfeature_speeds('Unknown')[0:]
#signal_l = butterfilter(old_data)

def LPC_plot(signal_l, C, signal_name):
    order = len(C)-1
    Cn = [] # C-negative
    for x1 in range(0,len(C)):
        Cn.append(C[x1]*-1)
    signal_lpc = []
    signal_diff = []
    for x1 in range(0,len(signal_l)-0):
        ytemp = []
        for x2 in range(0,len(C)-0):
            if -(x2+0)+x1 < 0:
                varde2 = 0
            else:
                varde = C[x2+0] * signal_l[0-(x2+0)+x1]
                varde2 = varde * -1
            ytemp.append(varde2)
        ynext = sum(ytemp)
        #if x1 < 2.5: # ignore some early results
        #    ynext = abs(signal_l[x1])
        signal_lpc.append(abs(ynext))
        signal_diff.append(signal_lpc[x1]-signal_l[x1])
    figure()
    plt.plot(range(0,len(signal_l)),signal_l[0:len(signal_l)], linewidth=1.0, linestyle = '-', color = 'k', label = "Original signal, filtered")
    plt.plot(range(order,len(signal_lpc)),signal_lpc[order:], linewidth=1.2, linestyle = '--',color = 'k', label = "LPC transformed signal")
    plt.plot(range(order,len(signal_diff)),signal_diff[order:], linewidth=0.7, linestyle = '-',color = 'k', label = "Difference")
    plt.subplots_adjust(right=2.0)
    grid()
    legend(bbox_to_anchor=(1.03, 1), loc=2, borderaxespad=0.)
    print 'Service:',signal_name, 'with order', order, 'has norm', linalg.norm(signal_diff), 'and meanSD =', meanSD(signal_diff)
    return signal_diff

def GETalpha(C):
    #print 'C',C
    Cmatrix = GETckmatrix(C)
    #print 'len C', len(C)
    #print 'len Cmatrix', len (Cmatrix)
    #print Cmatrix
    Cn = []
    for x1 in range(0,len(C)):
        Cn.append(C[x1]*-1)
    alpha = linalg.solve(Cmatrix[:-1,:-1],Cn[1:])
    #print 'len Cmatrix[:-1,:-1]', len(Cmatrix[:-1,:-1])
    #print 'len alpha',len(alpha)
    print 'alpha ', alpha
    return alpha

def LPC_plot2_procent(signal_l, C, signal_name):
    order = len(C)-0
    signal_lpc = []
    signal_diff = []
    for x1 in range(0,len(signal_l)-0):
        ytemp = []
        for x2 in range(0,order):
            if -(x2+0)+x1 < 0:
                varde2 = 0
            else:
                varde = C[x2+0] * signal_l[0-(x2+1)+x1]
                varde2 = varde * -1
            ytemp.append(varde2)
        ynext = sum(ytemp)
        if x1 < 2.5:
            ynext = abs(signal_l[x1])
        signal_lpc.append(abs(ynext))
        if (signal_lpc[x1]+signal_l[x1]) == 0:
            diffproc = 0
        else:
            diffproc = (signal_l[x1] - signal_lpc[x1]) / (signal_lpc[x1]+signal_l[x1])
        signal_diff.append(diffproc)
    figure()
    plt.plot(range(0,len(signal_l)),signal_l[:], linewidth=1.0, linestyle = '-', color = 'k', label = "Original signal, filtered")
    plt.plot(range(order,len(signal_lpc)),signal_lpc[order:len(signal_lpc)], linewidth=1.2, linestyle = '--',color = 'k', label = "LPC transformed signal")
    plt.subplots_adjust(right=2.0)
    legend(bbox_to_anchor=(1.03, 1), loc=2, borderaxespad=0.)
    figure()
    plt.plot(range(order,len(signal_diff)),signal_diff[order:len(signal_lpc)], linewidth=1, linestyle = '-',color = 'k', label = "Difference")
    plt.subplots_adjust(right=2.0)
    legend(bbox_to_anchor=(1.03, 1), loc=2, borderaxespad=0.)
    grid()
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    print 'Service:',signal_name, 'with order', order, 'has norm', linalg.norm(signal_diff), 'and meanSD =', meanSD(signal_diff)
    return signal_diff    
    
def MdiagADD(quantity, size):
    if (quantity+1)%2==0:
        print "good"
    else:
        print "Please enter odd integers"
        return 0
    y = numpy.zeros(size**2).reshape((size,size))
    v = []
    for x1 in range(0,quantity):
        v.append(1)
    r = -(quantity - 1) / 2
    for x1 in v:
        s = []
        for x2 in range(0,size-abs(r)):
            s.append(x1)
        y2 = np.diag(s,r)
        y = y + y2
        r = r + 1
    for x1 in range(0,len(y)):
        x3 = sum(y[x1])
        for x2 in range(0,len(y)):
            y[x1,x2] = y[x1,x2] / x3
    return y
```
